#4. Simulación estocástica primitiva

El siguiente programa simula la expresión de un gen regida por las siguientes ecuaciones

$$ \dot{r}(t) = k_r - \gamma_rr(t) $$
$$ \dot{p}(t) = k_pr(t) - \gamma_pp(t) $$

Donde utilizaremos la aproximación de $d t$ pequeños y tomaremos las constantes como $k_r = 1 \frac{\text{RNA}}{\text{min}}$, $k_p = 60 \frac{\text{proteinas}}{\text{RNA} \cdot \text{min}}$, $\gamma_r = \frac{1}{5} \frac{1}{\text{min}}$  y $\gamma_p = \frac{1}{30} \frac{1}{\text{min}}$. Los valores iniciales siempre serán $r(0)=0$ y $p(0)=0$.

```python
%pylab inline
import time

kr = 1.0 
gammar = 1.0/5.0 
kp = 60.0 
gammap = 1.0/30.0

r0 = 0
p0 = 0
```

La siguiente función realiza un paso de $dt$ y evalúa si en dicho intervalo hubo creación de RNA y/o de proteína. Recibe como parámetros la longitud del intervalo $dt$, el tiempo $t_{\text{old}}$, número de RNAs $r_{\text{old}}$ y número de proteínas $p_{\text{old}}$ del paso anterior y retorna los valores de estas cantidades para el intervalo actual a partir de ellas.

```python
def step(t_old, r_old, p_old, dt):
    
    t_new = t_old + dt
    rand_kr = random.random()
    rand_gammar = random.random()
    
    if(rand_kr < kr*dt and rand_gammar >= gammar*r_old*dt):
        r_new = r_old+1
    elif(rand_kr >= kr*dt and rand_gammar < gammar*r_old*dt):
        r_new = r_old-1
    else:
        r_new = r_old
    
    rand_kp = random.random()
    rand_gammap = random.random()
    
    if(rand_kp < kp*r_old*dt and rand_gammap >= gammap*p_old*dt):
        p_new = p_old+1
    elif(rand_kp >= kp*r_old*dt and rand_gammap < gammap*p_old*dt):
        p_new = p_old-1
    else:
        p_new = p_old
    
    return t_new, r_new, p_new
```

La siguiente función simula la regulación del gen en una sola célula durante un tiempo dado. Recibe como parámetros la longitud del intervalo $dt$ y el numero $n_{\text{points}}$ de intervalos de tiempo a ver, calculados a partir de $dt$ y el tiempo total $t_{\text{total}}$ durante el cual se va a ver el comportamiento de la celula. Retorna los arreglos de tiempo, numero de RNA y numero de proteínas como función del tiempo.

```python
def cell(dt, n_points):
    
    t = np.zeros(n_points)
    r = np.zeros(n_points)
    p = np.zeros(n_points)

    t[0] = 0
    r[0] = r0
    p[0] = p0
    
    for i in range(1, n_points):
        
        #cambiar la funcion llamada en la linea siguiente entre step y step_negative_feedback segun sea el caso
        t[i], r[i], p[i] = step_negative_feedback(t[i-1], r[i-1], p[i-1], dt)
    
    #Esta parte solo se incluye cuando miremos celulas individuales
    '''
    plot(t,r)
    plt.title("Number of RNA molecules of each cell", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of RNA molecules", fontsize = "13")
    plt.show()
    plt.close()
    
    plot(t, p)
    plt.title("Number of proteins of each cell", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of proteins", fontsize = "13")
    plt.show()
    plt.close()
    '''
    return t, r, p
```

Ejecutemos la simulación para una céula con un $t_{\text{total}} = 1000$ min y un $dt=0.0003$ min.

```python
t_total = 1000
dt = 0.0003
n_points = int(t_total/dt)

t, r, p = cell(dt, n_points)
```

De los resultados podemos ver que en estado estacionario el número de RNAs fluctúa mucho pero se mantiene alrededor de $5$, que es el valor de estado estacionario esperado según el modelo $r_{\text{ss}}=\frac{k_r}{\gamma_r} = 5$, e igualmente para el numero de proteinas, que se mantiene alrededor de 9000, que es su estado estacionario $p_{\text{ss}}=\frac{k_r k_p}{\gamma_r \gamma_p} = 9000$.

La escogencia de $dt = 0.0003$ min se hizo de tal forma que la probabilidad por intervalo se mantenga pequena siempre, particularmente del orden de $0.1$ y el razonamiento fue el siguiente: la probabilidad que se hará mas grande es la de degradación de proteínas ya que depende de su cantidad. Por lo tanto, si tenemos alrededor de $9000$ proteínas en estado estacionario, la probabilidad de degradación en un intervalo $dt$ es alrededor de $\gamma_p \cdot 9000 \cdot dt$. Si queremos que esto sea $0.1$, despejando $dt$ obtenemos.

$$dt = 0.00033 \: \text{min}$$

Es decir, del orden de tres diezmilésimas de minuto. El hecho de tener que tomar intervalos de tiempo tan pequeños hace que para lograr resultados válidos, el programa debe tomar mucho tiempo para ejecutarse como veremos más adelante.

La siguiente función realiza el procedimiento anterior para $N_{\text{cells}}$ células dado un tiempo $t_{\text{total}}$ durante el cual se va a estudiar el comportamiento del gen y la longitud del intervalo $dt$.

Esta función calcula y grafica: (1) Promedio de proteinas vs. tiempo, (2) Promedio de RNA vs. tiempo, (3) Ruido (CV) del numero de RNA vs. tiempo, (4) Ruido (CV) del numero de proteinas vs. tiempo, (5) Distribucion del numero RNA en el ultimo intervalo de tiempo, y (6) Distribucion del numero de proteinas en el ultimo intervalo de tiempo.

El promedio y el ruido, medido a través del coeficiente de variación se calculan para cada intervalo de tiempo sobre toda la población de células. El programa es ejecutado para un tiempo lo suficientemente largo para que se llegue a estado estacionario y las distribuciones en el último intervalo de tiempo sí correspondan a las distribuciones en estado estacionario.

Finalmente la función imprime el tiempo que tardó la ejecución del programa.

```python
def several_cells(N_cells, t_total, dt):
    
    start_time = time.time()
    
    n_points = int(t_total/dt)
    
    R = np.zeros((N_cells, n_points))
    P = np.zeros((N_cells, n_points))
    
    for i in range(N_cells):
        
        t, r, p = cell(dt, n_points)       
        R[i,:] = r
        P[i,:] = p
        
    ave_R = np.zeros(n_points)
    std_R = np.zeros(n_points)
    ave_P = np.zeros(n_points)
    std_P = np.zeros(n_points)
    
    for i in range(n_points):
        ave_R[i] = np.average(R[:,i])
        std_R[i] = np.std(R[:,i])
        ave_P[i] = np.average(P[:,i])
        std_P[i] = np.std(P[:,i])
       
    plot(t, ave_R)
    plt.title("Average of RNA molecules", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of RNA molecules", fontsize = "13")
    plt.show()
    plt.close()
    
    plot(t, ave_P)
    plt.title("Average of proteins", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of proteins", fontsize = "13")
    plt.show()
    plt.close()
    
    plot(t, std_R/ave_R)
    plt.title("Noise in RNA number", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("CV$_r$", fontsize = "13")
    ylim([0,5])
    plt.show()
    plt.close()
    
    plot(t, std_P/ave_P)
    plt.title("Noise in protein number", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("CV$_p$", fontsize = "13")
    ylim([0,5])
    plt.show()
    plt.close()
    
    hist(R[:, n_points-1], bins = 10)
    plt.title("Distribution of RNA in steady state", fontsize = "15")
    xlabel("Number of RNA molecules", fontsize = "13")
    ylabel("Number of cells", fontsize = "13")
    plt.show()
    plt.close()
       
    hist(P[:, n_points-1], bins = 60)
    plt.title("Distribution of proteins in steady state", fontsize = "15")
    xlabel("Number of proteins", fontsize = "13")
    ylabel("Number of cells", fontsize = "13")
    plt.show()
    plt.close()
    
    end_time = time.time() - start_time
    hours = int(end_time/3600)
    minutes = int((end_time%3600)/60)
    seconds = (end_time%3600)%60
    
    print("The execution time was: " + str(hours) + " hours, " + str(minutes) + " minutes, " + str(seconds) + " seconds.")
```

Dado que el programa toma mucho tiempo en ejecutar, *resulta muy dificil ver el comportamiento para $1000$ celulas y contar el tiempo para $10000$*. Entonces mediremos el tiempo de ejecucion y daremos nos resultados para $500$ celulas, con $dt = 0.0003$ min y $t_{\text{total}} = \frac{10}{\gamma_p} = 300$ min.

```python
N_cells = 500

t_total = 300
dt = 0.0003

several_cells(N_cells, t_total, dt)
```

En conclusión, para mantener una probabilidad máxima del orden de $0.1$ y que no cambien apreciablemente los resultados, el $dt$ debe ser del orden de una diezmilésima de minuto.

Ahora agregemos retroalimentacion negativa sobre la produccion de RNA, en este caso las ecuaciones quedan de la forma.

$$ \dot{r}(t) = \frac{k_r}{1+\left( \frac{p}{K} \right)^2} - \gamma_rr(t) $$
$$ \dot{p}(t) = k_pr(t) - \gamma_pp(t) $$

Para ello debemos definir el nuevo parametro $K$ que corresponde a la concentracion de proteínas a la cual se logra un $50$% de represión. Además se debe definir una nueva funcion de step que incluya la retroalimentación.

```python
K = 4500.0

def step_negative_feedback(t_old, r_old, p_old, dt):
    
    t_new = t_old + dt
    rand_kr = random.random()
    rand_gammar = random.random()
    
    (kr/(1.0 + (p_old/K)**2))
    
    if(rand_kr < (kr/(1.0 + (p_old/K)**2))*dt and rand_gammar >= gammar*r_old*dt):
        r_new = r_old+1
    elif(rand_kr >= (kr/(1.0 + (p_old/K)**2))*dt and rand_gammar < gammar*r_old*dt):
        r_new = r_old-1
    else:
        r_new = r_old
    
    rand_kp = random.random()
    rand_gammap = random.random()
    
    if(rand_kp < kp*r_old*dt and rand_gammap >= gammap*p_old*dt):
        p_new = p_old+1
    elif(rand_kp >= kp*r_old*dt and rand_gammap < gammap*p_old*dt):
        p_new = p_old-1
    else:
        p_new = p_old
    
    return t_new, r_new, p_new
```

Definimos $K = 4500$ proteinas, que corresponde a la mitad del numero de proteinas de estado estacionario cuando no hay retroalimentacion.

Como vimos en clase, se esperaría que bajo retroalimentación negativa, el número de proteínas en estado estacionario es exactamente $K$. Además, el tiempo de respuesta del sistema es mucho menor y por lo tanto se alcanza más rápido el estado estacionario con respecto al caso donde no hay regulación.

```python
t_total = 1000
dt = 0.0003
n_points = int(t_total/dt)

t, r, p = cell(dt, n_points)
```

Podemos ver que para las proteínas, efectivamente se alcanza el estado estacionario a alrededor de $4500$ y luego el sistema fluctua alrededor de ese valor, siendo reprimido cuando la cantidad de proteína aumenta mucho y pudiendo aumentar la cantidad cuando disminuye.

Ahora ejecutemos el algoritmo para retroalimentación negativa con $500$ celulas, con $dt = 0.0003$ min y $t_{\text{total}} = \frac{10}{\gamma_p} = 300$ min

Las proteinas sí alcanzan el valor de estado estacionario de $K = 4500$ como se esperaba. Para el RNA también hay un crecimiento más acelerado antes de llegar al estado estacionario que en el caso donde no hay autorregulación, alcanzando por primera vez su valor de estado estacionario en menos de $10$ min (antes de estabilizarse), mientras que en el caso sin regulación este tiempo es de alrededor de $20$ min.
