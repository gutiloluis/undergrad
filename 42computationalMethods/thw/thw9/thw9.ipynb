```python
countA = []
countG = []
countT = []
countC = []

infile = open("cromX.dat", "r")
data = infile.readlines()


for line in data:
    countA.append(line.count('A'))
    countG.append(line.count('G'))
    countT.append(line.count('T'))
    countC.append(line.count('C'))
```

```python
print("promedio de A: " + str(mean(countA)))
print("promedio de G: " + str(mean(countG)))
print("promedio de T: " + str(mean(countT)))
print("promedio de C: " + str(mean(countC)))
```

```python
print("desviacion estandar de A: " + str(std(countA)))
print("desviacion estandar de G: " + str(std(countG)))
print("desviacion estandar de T: " + str(std(countT)))
print("desviacion estandar de C: " + str(std(countC)))
```

```python
print("mediana de A: " + str(median(countA)))
print("mediana de G: " + str(median(countG)))
print("mediana de T: " + str(median(countT)))
print("mediana de C: " + str(median(countC)))
```

```python
# Two subplots, the axes array is 1-d
figure(figsize(15,15))
f, axarr = plt.subplots(3, sharex=True)
axarr[0].set_title("Estudio de la frecuencia de las bases nitrogenadas\n en fragmentos del cromosoma X", fontsize = 18)

plt.xlim(0,40)
plt.xlabel("$N$", fontsize = 18)
axarr[1].set_ylabel("$f_a$", fontsize = 18)

h1 = axarr[0].hist([countG,countC], normed = True, bins = 50, cumulative = True, label = ['Guanina', 'Citosina'],color = ['b','r'])
axarr[0].legend()
h2 = axarr[1].hist([countA,countT], normed = True, bins = 50, cumulative = True, label = ['Adenina', 'Timina'],color = ['y','black'])
axarr[1].legend()
h3 = axarr[2].hist([countG,countA], normed = True, bins = 50, cumulative = True, label = ['Guanina', 'Adenina'],color = ['b','y'])
axarr[2].legend()
```
