Ornstein-Uhlenbeck process

```python
def step(t_old, x_old):
    
    t_new = t_old + dt
    x_new = x_old + theta*(mu-x_old)*dt + sigma*np.sqrt(dt)*np.random.normal()
    
    return t_new, x_new
```

```python
t_total = 10.0
mu = 1.0
sigma = 0.1
theta = 1.0
x_0 = 5.0
dt = 0.001
n_points = int(t_total/dt)
    
def steps():

    T = np.zeros(n_points)
    X = np.zeros(n_points)

    T[0] = 0
    X[0] = x_0

    for i in range(1,n_points):
    
        t_new, x_new = step(T[i-1],X[i-1])
        T[i] = t_new
        X[i] = x_new
        
    return T, X
```

```python
n_runs = 20

TT = np.zeros((n_points,n_runs))
XX = np.zeros((n_points,n_runs))

for i in range(n_runs):
    T,X = steps()
    TT[:,i] = T
    XX[:,i] = X
```

```python
X_ave = np.zeros(n_points)
X_var = np.zeros(n_points)
for i in range(n_points):
    X_ave[i] = np.average(XX[i,:])
    X_var[i] = np.var(XX[i,:])
```

```python
plot(T,X_ave)
```

```python
plot(T,X_var)
```
