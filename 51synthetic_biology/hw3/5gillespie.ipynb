#5. Simulación estocástica con algoritmo de Gillespie

Modelaremos las mismas ecuaciones con las mismas constantes y condiciones iniciales, y seguiremos los mismos pasos que en el ejercicio anterior, pero ahora utilizaremos el algoritmo de Gillespie. El análisis y las explicaciones se realizaron detalladamente en el punto anterior, por lo tanto aquí no nos centraremos en ello.

```python
%pylab inline
import time

kr = 1.0 
gammar = 1.0/5.0 
kp = 60.0
gammap = 1.0/30.0

r0 = 0
p0 = 0
```

En la función step, a diferencia del caso primitivo, no se define un intervalo de tiempo ya que los saltos de tiempo son definidos a partir de un número aleatorio con una distribución exponencial. Otro número aleatorio se utiliza para definir cual de los eventos (creación de RNA, destrucción de RNA, creación de proteína o destrucción de proteína) ocurrió.

```python
def step(t_old, r_old, p_old):
    
    K = kr + gammar*r_old + kp*r_old + gammap*p_old
    
    Dt = np.log(1.0/random.random())/K
    which = np.random.random()

    t_new = t_old + Dt
    
    if(which < kr/K):
        r_new = r_old + 1
        p_new = p_old
    elif(which >= kr/K and which < (kr+gammar*r_old)/K):
        r_new = r_old - 1
        p_new = p_old
    elif(which >= (kr+gammar*r_old)/K and which < (kr+gammar*r_old+kp*r_old)/K):
        r_new = r_old
        p_new = p_old + 1
    else:
        r_new = r_old
        p_new = p_old - 1
        
    return t_new, r_new, p_new
```

En la función cell, t$_c$ denota el tiempo correspondiente al paso actual y el algoritmo continúa ejecutándose hasta que dicho tiempo alcance $t_{\text{total}}$. Además esta función reescala los arreglos $t$, $r$ y $p$ en $t_s$, $r_s$ y $p_s$ de tal forma que tengan la misma escala de tiempo y se puedan realizar promedios sobre las células. Aquí $dt$ denota el intervalo de tiempo con el que se van a reescalar los arrays. La funcion retorna los arrays reescalados.

```python
def cell(dt, t_total):
    
    t = []
    r = []
    p = []
    
    t_c = 0
    r_c = r0
    p_c = p0
    
    t.append(0)
    r.append(r0)
    p.append(p0)
    
    while t_c < t_total:
        
        #cambiar la funcion llamada en la linea siguiente entre step y step_negative_feedback segun sea el caso
        t_c, r_c, p_c = step(t_c, r_c, p_c)
        
        t.append(t_c)
        r.append(r_c)
        p.append(p_c)
    
    t_s = np.linspace(0, t_total-dt, t_total/dt)
    r_s = np.zeros(len(t_s))
    p_s = np.zeros(len(t_s))
    
    j = 0
    for i in range(len(t_s)):
        
        while(t[j] < t_s[i]):
            j+=1
            
        r_s[i] = r[j]
        p_s[i] = p[j]
        
    #Esta parte solo se incluye cuando miremos celulas individuales
    '''
    plot(t,r)
    plt.title("Number of RNA molecules of each cell", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of RNA molecules", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    
    plot(t_s,r_s)
    plt.title("Number of RNA molecules of each cell (scaled)", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of RNA molecules", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    
    plot(t, p)
    plt.title("Number of proteins of each cell", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of proteins", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    
    plot(t_s, p_s)
    plt.title("Number of proteins of each cell (scaled)", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of proteins", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    '''
    
    return t_s, r_s, p_s
```

Corramos el algoritmo para $t_{\text{total}} = 1000$ min y $dt = 0.001$ min.

```python
t_total = 1000
dt = 0.001

t_s, r_s, p_s = cell(dt, t_total)
```

Se hicieron, tanto para RNA como para proteínas, de los arrays sin escalar y escalados y podemos ver que con el $dt$ que tomamos no perdemos mucha resolución pues no hay diferencias apreciables. Verificamos también que el algoritmo funciona al comparar los resultados con el punto anterior.

La siguiente funcion hace exactamente lo mismo que la del punto anterior con los arrays reescalados.

```python
def several_cells(N_cells, t_total, dt):
    
    start_time = time.time()
    
    n_points = int(t_total/dt)
    
    R = np.zeros((N_cells, n_points))
    P = np.zeros((N_cells, n_points))
    
    for i in range(N_cells):
        
        t_s, r_s, p_s = cell(dt, t_total)
        
        R[i,:] = r_s
        P[i,:] = p_s
         
    ave_R = np.zeros(n_points)
    std_R = np.zeros(n_points)
    ave_P = np.zeros(n_points)
    std_P = np.zeros(n_points)
    
    for i in range(n_points):
        ave_R[i] = np.average(R[:,i])
        std_R[i] = np.std(R[:,i])
        ave_P[i] = np.average(P[:,i])
        std_P[i] = np.std(P[:,i])
   
    plot(t_s, ave_R)
    plt.title("Average of RNA molecules", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of RNA molecules", fontsize = "13")
    plt.show()
    plt.close()
    
    plot(t_s, ave_P)
    plt.title("Average of proteins", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("Number of proteins", fontsize = "13")
    ylim([0,10000])
    plt.show()
    plt.close()
    
    plot(t_s, std_R/ave_R)
    plt.title("Noise in RNA number", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("CV$_r$", fontsize = "13")
    ylim([0,2.5])
    plt.show()
    plt.close()
    
    plot(t_s, std_P/ave_P)
    plt.title("Noise in protein number", fontsize = "15")
    xlabel("time (minutes)", fontsize = "13")
    ylabel("CV$_p$", fontsize = "13")
    ylim([0,2.5])
    plt.show()
    plt.close()
    
    hist(R[:, n_points-1], bins = 8)
    plt.title("Distribution of RNA in steady state", fontsize = "15")
    xlabel("Number of RNA molecules", fontsize = "13")
    ylabel("Number of cells", fontsize = "13")
    plt.show()
    plt.close()
    
    hist(P[:, n_points-1], bins = 60)
    plt.title("Distribution of proteins in steady state", fontsize = "15")
    xlabel("Number of proteins", fontsize = "13")
    ylabel("Number of cells", fontsize = "13")
    plt.show()
    plt.close()
    
    end_time = time.time() - start_time
    hours = int(end_time/3600)
    minutes = int((end_time%3600)/60)
    seconds = (end_time%3600)%60
    
    print("The execution time was: " + str(hours) + " hours, " + str(minutes) + " minutes, " + str(seconds) + " seconds.")
```

Ahora ejecutemos la función anterior con las mismas condiciones en las que se realizó con la simulación primitiva. $N_{\text{cells}} = 500$ y $t_{\text{total}} = \frac{10}{\gamma_p} = 300$ min. El intervalo de tiempo de escalamiento es de $0.001$ min.

```python
N_cells = 500
t_total = 300
dt = 0.001

several_cells(N_cells, t_total, dt)
```

Como podemos notar, el tiempo de ejecución es considerablemente menor que el de la simulación primitiva, que fue de 1 hora, 30 minutos y 40.3723700047 segundos; mientras que con Gillespie fue de 15 minutos y 17.1972210407 segundos, es decir, usando Gillespie la ejecución tarde una sexta parte de lo que tarda empleando el algoritmo primitivo.

Si incluimos retroalimentacion negativa debemos modificar la funcion step para incluirla y definir el parametro $K$, tomamos el mismo valor que en el punto anterior.

```python
k = 4500.0

def step_negative_feedback(t_old, r_old, p_old):
    
    Kr = kr/(1.0 + (p_old/k)**2)
    
    K = Kr + gammar*r_old + kp*r_old + gammap*p_old
    
    Dt = np.log(1.0/random.random())/K
    which = np.random.random()
    t_new = t_old + Dt
    
    if(which < Kr/K):
        r_new = r_old + 1
        p_new = p_old
    elif(which >= Kr/K and which < (Kr+gammar*r_old)/K):
        r_new = r_old - 1
        p_new = p_old
    elif(which >= (Kr+gammar*r_old)/K and which < (Kr+gammar*r_old+kp*r_old)/K):
        r_new = r_old
        p_new = p_old + 1
    else:
        r_new = r_old
        p_new = p_old - 1
        
    return t_new, r_new, p_new
```

Corriendo el algoritmo para una célula a $1000$ min obtenemos.

```python
t_total = 1000
dt = 0.001

t_s, r_s, p_s = cell(dt, t_total)
```

Vemos que también obtenemos los resultados esperados según lo dicho en el punto anterior y vemos que no es notoria la información perdida al reescalar los arrays.

Ahora ejecutemos el algoritmo para $500$ celulas, con $dt = 0.001$ min y $t_{\text{total}} = \frac{10}{\gamma_p} = 300$ min

```python
N_cells = 500

t_total = 300
dt = 0.001

several_cells(N_cells, t_total, dt)
```

El tiempo de ejecución fue de 9 minutos y 55 segundos, en comparación con el tiempo de ejecución del algoritmo primitivo que fue de 2 horas, 27 minutos y 6 segundos para retroalimentación negativa. Aquí el algoritmo primitivo tardó aproximadamente 15 veces más con los mismos parámetros.

Tanto con retroalimentación como sin ella, el tiempo de ejecución del algoritmo de Gillespie es mucho menor. Además, este último ofrece la ventaja de que (salvo por el reescalamiento para comparar entre varias células), no depende de un intervalo entre los eventos, debido a esto es mucho más preciso que el algoritmo primitivo.
