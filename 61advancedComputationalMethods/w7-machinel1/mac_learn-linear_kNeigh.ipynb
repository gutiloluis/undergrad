```python
%pylab inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap
```

```python
def set_points(n_points):

    mean_blue_ch = np.array([0,1])
    covariance_ch = np.array([[1,0],[0,1]])
    mean_orange_ch = np.array([1,0])

    blue_ch = np.random.multivariate_normal(mean_blue_ch, covariance_ch, 10)
    orange_ch = np.random.multivariate_normal(mean_orange_ch, covariance_ch, 10)
    
    blue = np.zeros((n_points,2))
    orange = np.zeros((n_points,2))

    for i in range(n_points):
        mean_blue = [np.random.choice(blue_ch[:,0]),np.random.choice(blue_ch[:,1])]
        mean_orange = [np.random.choice(orange_ch[:,0]),np.random.choice(orange_ch[:,1])]
        blue[i,:] = np.random.multivariate_normal(mean_blue,1.0/5.0*covariance_ch)
        orange[i,:] = np.random.multivariate_normal(mean_orange,1.0/5.0*covariance_ch)
        
    X = np.concatenate((blue,orange))
    y = np.concatenate((np.zeros(n_points),np.ones(n_points)))    

    plt.scatter(blue[:,0],blue[:,1], c ='blue')
    plt.scatter(orange[:,0],orange[:,1], c='orange')
    plt.xlim(X[:,0].min()-1,X[:,0].max()+1)
    plt.ylim(X[:,1].min()-1,X[:,1].max()+1)
        
    return X, y
```

```python
n_points = 200
X, y = set_points(n_points)
```

```python
def linear_regression(X, y):

    regr = LinearRegression()
    regr.fit(X,y)

    h = 0.02

    cmap_light = ListedColormap(['#7E78FA', '#FAC478'])
    cmap_bold = ListedColormap(['#210AF2', '#F2C40A'])

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = regr.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-Class classification (linear regression)")
```

```python
linear_regression(X,y)
```

```python
def k_nearest_neighbors(X, y, n_neighbors, weights):
    
    clf = KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X,y)

    h = 0.02

    cmap_light = ListedColormap(['#7E78FA', '#FAC478'])
    cmap_bold = ListedColormap(['#210AF2', '#F2C40A'])

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-Class classification (k = %i, weights = '%s')" %(n_neighbors, weights))
    
```

```python
k_nearest_neighbors(X, y, 15, 'uniform')
```

```python
k_nearest_neighbors(X, y, 1, 'distance')
```
