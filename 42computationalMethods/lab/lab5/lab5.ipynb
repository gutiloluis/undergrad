```python
%pylab inline
from scipy.fftpack import fft, fftfreq
```

```python
n_points  = 1024
rand_signal  = random.random(n_points) 
rand_signal = rand_signal -  np.mean(rand_signal)
plot(rand_signal)
title("Random signal")
```

```python
fft_rand_signal = fft(rand_signal) / n_points 
freq = fftfreq(n_points, 1) 
plt.plot(np.fft.fftshift(freq),np.fft.fftshift(np.abs(fft_rand_signal)))
title("Fourier Transform of the random signal")
```

```python
power_spectrum = abs(fft_rand_signal)**2
plot(np.fft.fftshift(freq), np.fft.fftshift(power_spectrum))
title("Power spectrum of the random signal")
```

```python
fft_white_noise = []
for element in fft_rand_signal:
    fft_white_noise.append(element/abs(element))
    
plot(np.fft.fftshift(freq),np.fft.fftshift((np.abs(fft_white_noise)**2)))
title("Normalized power spectrum")
```

```python
white_noise = ifft(fft_white_noise)
figure(figsize(13,7))
plot(rand_signal, label = "Random signal")
plot(white_noise, label = "White noise")
legend()
title("White noise and random signal", fontsize = 15)
xlim(0,1024)
```
