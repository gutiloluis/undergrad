###Cellular automata

```python
#generates the pattern on the ith row of the matrix of M columns according to rule 90
def generate_row_rule_90(prv_row, M):
    
    row = zeros(M)
    
    for j in range(M):
        
        if j!=0 and j!=M-1:
            prv = j-1
            nxt = j+1
        elif j == 0:
            prv = M-1
            nxt = j+1
        else:
            prv = j-1
            nxt = 0

        iz = prv_row[prv]
        ce = prv_row[j]
        de = prv_row[nxt]
        
        
        if prv_row[prv] == 0 and prv_row[j] == 0 and prv_row[nxt] == 0:
            row[j] = 1
        elif prv_row[prv] == 0 and prv_row[j] == 0 and prv_row[nxt] == 1:
            row[j] = 0
        elif prv_row[prv] == 0 and prv_row[j] == 1 and prv_row[nxt] == 0:
            row[j] = 1
        elif prv_row[prv] == 0 and prv_row[j] == 1 and prv_row[nxt] == 1:
            row[j] = 0
        elif prv_row[prv] == 1 and prv_row[j] == 0 and prv_row[nxt] == 0:
            row[j] = 0
        elif prv_row[prv] == 1 and prv_row[j] == 0 and prv_row[nxt] == 1:
            row[j] = 1
        elif prv_row[prv] == 1 and prv_row[j] == 1 and prv_row[nxt] == 0:
            row[j] = 0
        elif prv_row[prv] == 1 and prv_row[j] == 1 and prv_row[nxt] == 1:
            row[j] = 1
            
    return row
```

```python
M = 500
N = 250
matrix = np.zeros((N,M))

matrix[0,250] = 1
for i in range(1,N):
    
    prv_row = matrix[i-1,:]
    matrix[i,:] = generate_row_rule_90(prv_row, M)

figure(figsize(16,8))
imshow(matrix, cmap='gray')
```

### Random Walk

```python
def random_walk(N):
    
    x = np.zeros(N+1)
    y = np.zeros(N+1)
    
    x[0] = 0
    y[0] = 0
    
    for i in range(1,N+1):
        
        theta = 2*pi*random.random()
        dx = cos(theta)
        dy = sin(theta)
        
        x[i] = x[i-1]+dx
        y[i] = y[i-1]+dy
    
    return x, y
```

```python
figure(figsize=(12,12))
x, y = random_walk(300)
xlabel("$x$", fontsize = 20)
ylabel("$y$", fontsize = 20)
plot(x,y)
scatter(x,y,color = 'g')
ax = plt.axes()
arr = ax.arrow(0, 0, x[300], y[300],width = 0.08, head_width=0.3, head_length=0.5, fc='k', ec='k')
```
