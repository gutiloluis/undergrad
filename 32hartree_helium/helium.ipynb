###Analisis físico del resultado

El valor de energía obtenido corresponde al menos primera energía de ionización del átomo de Helio, que sedefine como la energía necesaria para remover un electrón del átomo en estado base. El valor según laliteratura es de $24.6$ eV y por ende tiene todo el sentido que el electrón tenga energía que teóricamentees $-24.6$ eV (y que se desvió en nuestros resultados por aproximadamente 3 décimas), ya que si éste electrón adquiere una energía de $24.6$ eV el electron queda libre con $E = 0$ y el átomo queda ionizado.Ahora bien, luego de retirarse el primer electrón el átomo de Helio ya es un átomo hidrogenoide con energía $E = -Z^2 \cdot 13.6 $ eV = $ -54.4 $ eV. Debemos notar con alegría lo eficaz y bonito que eseste método ya que a partir de varias aproximaciones, entre ellas el hecho de tratar el movimiento de cadaelectrón de manera independiente a pesar de considerar las fuerzas de Coulomb entre ellos permitiendo separarla ecuación de Schrodinger para el sistema de manera independiente para cada electrón, y además de la interpolación de una función de potencial de prueba teniendo en cuenta sólolas condiciones para $u$ grande y pequeño e ir construyendo a partir de allí una función de potencial muy compatible con el valor de la energía que esperábamos encontrar.La diferencia entre el valor esperado y el valor que se obtuvo con el algoritmo se justifica debido alerror computacional, que aunque se intentó reducir al máximo siempre afectará ligeramente los resultados.También a que es difícil ver con extrema precisión la convergencia de las funciones logarítmicas entoncesno hay completa precisión en este aspecto y cambios de décimas y centésimas alteran los resultados enormemente.
