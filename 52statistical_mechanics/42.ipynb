Las siguientes funciones realizan el modelo utilizando el algoritmo de Gillespie. Hay que tener en cuenta que el tiempo está re-escalado desde 0 hasta 600. Así, t = 0 corresponde a las 9 a.m. y t = 600 a las 7 p.m. Además, t = 480 corresponde a las 5 p.m., que es cuando cambia la tasa de llegada de clientes.

Para hacer la conversión se realiza:

$$t_r = t_p + 540 \text{ mins}$$

Donde $t_r$ es el tiempo real y t_p es el tiempo del programa, ambos en minutos.

```python
def llega_cliente(Fila, tiempo):
    
    Fila.append(tiempo)

def atiende_cliente(Fila, tiempo):
    
    t_espera = tiempo - Fila[0]
    Fila.pop(0)
    
    return t_espera

#Como es esta?
def aburre_cliente(Fila, tiempo):
    
    t_espera = tiempo - Fila[0]
    Fila.pop(-1)
    
    return t_espera
```

```python
arr_1 = 1.0/5.0 #1/min (Tasa de llegada entre 9 y 5)
arr_2 = 1.0/2.0 #1/min (Tasa de llegada entre 5 y 7)
assi = 1.0/2.0 #1/min (Tasa de atencion)
bor = 1.0/10.0 #1/min (Tasa de aburrimiento)

p0 = 0
p_lost0 = 0

def paso(t_old, p_old, p_atend_old, p_lost_old, t_espera_total_old, Fila):
    
    K = 0.0
    Dt = np.log(1.0/random.random())
    which = np.random.random()
    
    p_new = p_old
    p_atend_new = p_atend_old
    p_lost_new = p_lost_old
    t_espera_total_new = t_espera_total_old
    
    #print "NEW"
    #entre 9 y 5
    if(t_old < 480):
        
        #print "entre 9 y 5"
        K += arr_1
         
        #Si hay clientes
        if(p_old > 0):
            
            #print "  hay clientes"
            K += assi
               
            #Si hay mas de 10
            if(p_old > 10):              
                #print "    hay mas de 10"
                K += bor                
                Dt/=K
                t_new = t_old + Dt
                #En este caso llega un cliente
                if(which < arr_1/K):
                    p_new+=1
                    llega_cliente(Fila, t_new)
                #En este caso se atiende un cliente
                elif(which >= arr_1/K and which < (arr_1 + assi)/K):
                    p_new-=1
                    p_atend_new+=1
                    t_espera_total_new+=atiende_cliente(Fila, t_new)
                #En este caso se aburre un cliente
                else:
                    #OJOOOO AQUIIIIII SI ESTE TIEMPO TAMBIEN SE INCLUYE EN EL PROMEDIO
                    p_new-=1
                    p_lost_new+=1
                    t_espera_total_new+=aburre_cliente(Fila, t_new)
                       
            #Si no hay mas de 10
            else:               
                #print "    no hay mas de 10"
                Dt/=K
                t_new = t_old + Dt
                #En este caso llega un cliente
                if(which < arr_1/K):
                    p_new+=1
                    llega_cliente(Fila, t_new)
                #En este caso se atiende un cliente
                else:
                    p_new-=1
                    p_atend_new+=1
                    t_espera_total_new+=atiende_cliente(Fila, t_new)
                    
        #Si no hay clientes
        else:        
            #print "  no hay clientes"
            Dt/=K
            t_new = t_old + Dt
            #En este caso llega un cliente
            p_new+=1
            llega_cliente(Fila, t_new)
                               
    #entre 5 y 7    
    else:           
        K += arr_2
        
        #Si hay clientes
        if(p_old > 0):   
            K += assi
            
            #Si hay mas de 10
            if(p_old > 10):        
                K += bor                
                Dt/=K
                t_new = t_old + Dt
                #En este caso llega un cliente
                if(which < arr_2/K):
                    p_new+=1
                    llega_cliente(Fila, t_new)
                #En este caso se atiende un cliente
                elif(which >= arr_2/K and which < (arr_2 + assi)/K):
                    p_new-=1
                    p_atend_new+=1
                    t_espera_total_new+=atiende_cliente(Fila, t_new)
                #En este caso se aburre un cliente
                else:
                    p_new-=1
                    p_lost_new+=1
                    t_espera_total_new+=aburre_cliente(Fila, t_new)
            
            #Si no hay mas de 10
            else:
                
                Dt/=K
                t_new = t_old + Dt
                #En este caso llega un cliente
                if(which < arr_2/K):
                    p_new+=1
                    llega_cliente(Fila, t_new)
                #En este caso se atiende un cliente
                else:
                    p_new-=1
                    p_atend_new+=1
                    t_espera_total_new+=atiende_cliente(Fila, t_new)
        #Si no hay
        else:
            
            Dt/=K
            t_new = t_old + Dt
            #En este caso llega un cliente
            p_new+=1
            llega_cliente(Fila, t_new)
                
    return t_new, p_new, p_atend_new, p_lost_new, t_espera_total_new
```

```python
def cajero(dt, t_total):
    
    t = []
    p = []
    p_atend = []
    p_lost = []
    t_espera_total = []
    
    Fila = []
    
    t_c = 0
    p_c = p0
    p_atend_c = 0
    p_lost_c = p_lost0
    t_espera_total_c = 0
    
    t.append(0)
    p.append(p0)
    p_atend.append(0)
    p_lost.append(p_lost0)
    t_espera_total.append(0)
    
    while t_c < t_total:
        
        t_c, p_c, p_atend_c, p_lost_c, t_espera_total_c = paso(t_c, p_c, p_atend_c, p_lost_c, t_espera_total_c, Fila)
        
        t.append(t_c)
        p.append(p_c)
        p_atend.append(p_atend_c)
        p_lost.append(p_lost_c)
        t_espera_total.append(t_espera_total_c)
    
    t_espera_prom = 1.0*np.array(t_espera_total)/(1.0*np.array(p_atend) + 1.0*np.array(p_lost))
    
    #Esta parte solo se incluye cuando miremos una corrida individual
    
    '''
    plot(t,p)
    plt.title("Length of queue", fontsize = "15")
    xlabel("Time (minutes)", fontsize = "13")
    ylabel("Number of people", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
  
    plot(t,p_atend)
    plt.title("Assisted customers", fontsize = "15")
    xlabel("Time (minutes)", fontsize = "13")
    ylabel("Number of people", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    
    plot(t, p_lost)
    plt.title("Lost customers", fontsize = "15")
    xlabel("Time (minutes)", fontsize = "13")
    ylabel("Number of people", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    
    plot(t, t_espera_prom)
    plt.title("Average waiting time", fontsize = "15")
    xlabel("Time (minutes)", fontsize = "13")
    ylabel("Waiting time (minutes)", fontsize = "13")
    xlim([0,t_total])
    plt.show()
    plt.close()
    '''
    
    return t, p, p_lost, t_espera_prom  
```

Las siguientes funciones realizan corridas de varios dias

```python
def cajeros(N, dt, t_total):
    
    T = zeros(N)
    
    for i in range(N):
        
        t, p, p_lost, t_espera_prom = cajero(dt, t_total)
        T[i] = t_espera_prom[-1]
    
    return T
```

```python
t_total = 600
dt = 0.01
N = 3000

T = cajeros(N, dt, t_total)
```

```python
def frequencies(Array):
    B = []
    C = []
    for el in Array:
        if el not in B:
            B.append(el)
            C.append(1)
        elif el in B:
            C[B.index(el)]+=1
    return B, C
```

```python
T1,T2 = frequencies(T)
```

A continuación se muestra el histograma para el tiempo de espera promedio en un día, y las estadísticas (promedio, moda y desviación estándar (STD)). Se corrieron 3000 dias.

Las siguientes funciones hacen corridas por meses y van calculando el número de clientes perdidos

```python
def month():

    p_lost_m = 0

    for i in range(30):
        dt = 0.1
        t_total = 600
        t, p, p_lost, t_espera_prom = cajero(dt, t_total)
        p_lost_m += p_lost[-1]
    
    return p_lost_m
```

```python
def several_months(N):
    
    P_Lost_M = zeros(N)
    
    for i in range(N):
        
        P_Lost_M[i] = month()
    
    return P_Lost_M
```

```python
P_Lost_M = several_months(1000)
P1,P2 = frequencies(P_Lost_M)
```

A continuación se muestra el histograma para el número de clientes perdidos mensualmente, y las estadísticas (promedio, moda y desviación estándar (STD)).
