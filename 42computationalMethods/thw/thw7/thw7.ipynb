```python
def f(z):
    return z**3 - 1

def f_prime(z):
    return 3*z**2
```

```python
def NR(guess_0):
    guess = guess_0
    for i in range(10):
        delta_x = -f(guess)/f_prime(guess)
        guess = guess + delta_x
    return guess
```

```python
def closer_to(z):
    d = abs(z-1)
    num = 1
    
    if(abs((-0.5 + (sqrt(3)/2.0)*1j)-z)<d):
        num = 2
        d = abs((-0.5 + (sqrt(3)/2.0)*1j)-z)
        
    if(abs((-0.5 - (sqrt(3)/2.0)*1j)-z)<d):
        num = 3
        
    return num
```

```python
arr = zeros((500,500), dtype = complex)
re = linspace(-1,1,500)
im = linspace(1j,-1j,500)
for i in range(500):
    for j in range(500):
        arr[i,j]=re[i]+im[j]
```

```python
arr2 = zeros((500,500))
for i in range(500):
    for j in range(500):
        arr2[i,j] = closer_to(NR(arr[i,j]))
```

```python
figure(figsize(10,10))
imshow(arr2)
```
