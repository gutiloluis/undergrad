```python
%pylab inline
from scipy import fftpack
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
Im1 = pylab.imread("rectangular.gif")
fig = figure(1, figsize=(10.5,8.5))
imshow(Im1[:,:], cmap = cm.Greys_r)
plt.title('20_popc_cho009-1.tif', fontsize=12)
```

```python
Im1_ft = fftpack.fft2(Im1, axes=(0, 1))
```

```python
n_x = shape(Im1)[1]
n_y = shape(Im1)[0]

middle_x = shape(Im1)[1]/2.0
middle_y = shape(Im1)[0]/2.0

#m
number_x = shape(Im1)[1]/1.2
number_y = shape(Im1)[0]/1.2

low_x = int(middle_x - number_x/2.0)
up_x = int(middle_x + number_x/2.0)

low_y = int(middle_y - number_y/2.0)
up_y = int(middle_y + number_y/2.0)

print(middle_x,middle_y)
print(number_x,number_y)
print(low_x,up_x)
print(low_y,up_y)
```

```python
for i in range(n_x):
    for j in range(n_y):
        if ((i > low_x) and (i < up_x) and (j > low_y) and (j < up_y)):
            Im1_ft[j,i] = 0.0
```

```python
low_Im1 = fftpack.ifft2(Im1_ft, axes=(0, 1)).real
```

```python
fig = figure(1, figsize=(10.5,8.5))
imshow(low_Im1[:,:], cmap = cm.Greys_r)
plt.title('20_popc_cho009-1.tif', fontsize=12)
```
