##a. $\int_{0}^{\pi} \sin^2(\pi \cos(\theta)) \cos^2(\theta) d\theta$

```python
def f_a(x):
    return (sin(pi*cos(x)))**2 * (cos(x))**2

min_x = 0.0
max_x = pi

x = linspace(min_x,max_x,1000)

plot(x, f_a(x))
xlabel("$\\theta$", fontsize = 18)
ylabel("$f_a(\\theta)$", fontsize = 18)
```

```python
min_y = amin(f_a(x))
max_y = amax(f_a(x))

n_random = 20000
random_x = random.rand(n_random) * (max_x - min_x) + min_x
random_y = random.rand(n_random) * (max_y - min_y) + min_y

delta = f_a(random_x) - random_y
below  = where(delta>0.0)

scatter(random_x, random_y)
scatter(random_x[below], random_y[below], c = 'r')
xlabel("$\\theta$", fontsize = 18)
ylabel("$f_a(\\theta)$", fontsize = 18)
xlim((min_x, max_x))
ylim((min_y, max_y))

integral = (max_y - min_y)*(max_x - min_x)*size(below)/size(random_x)
print "The integral using Monte-Carlo is: " + str(integral) + " \nAnd the integral using quad with its numerical error is " + str(quad(f_a, min_x, max_x))
```

##f. $\int_{0}^{\infty} e^{-(x^2+y^2+z^2+w^2)} dxdydzdw = \pi^2/16 = 0.6168$

```python
def f_f(x1, x2, x3, x4):
    return exp(-(x1**2 + x2**2 + x3**2 + x4**2))

min_x = 0
max_x = 10.0

min_y = 0.0
max_y = 1.0

n_random = 20000000
random_x1 = random.rand(n_random) * (max_x - min_x) + min_x
random_x2 = random.rand(n_random) * (max_x - min_x) + min_x
random_x3 = random.rand(n_random) * (max_x - min_x) + min_x
random_x4 = random.rand(n_random) * (max_x - min_x) + min_x

random_y = random.rand(n_random) * (max_y - min_y) + min_y

delta = f_f(random_x1, random_x2, random_x3, random_x4) - random_y

below  = where(delta>0.0)

integral = (max_y - min_y)*((max_x - min_x)**4)*size(below)/size(random_x1)

print "The integral using Monte-Carlo is: " + str(integral)
```
