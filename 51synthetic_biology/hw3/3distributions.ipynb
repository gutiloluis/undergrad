###3. Distribuciones de probabilidad

A continuación se grafican algunas distribuciones de probabilidad para distintos valores de sus respectivos parámetros.

```python
%pylab inline
from scipy.stats import *
```

####a. Poisson

```python
x = np.linspace(0, 30, 31)
figure(figsize=(5, 3.6))
scatter(x,poisson.pmf(x, 10))
ylabel("$f_X(x)$", fontsize = "18")
xlabel("$x$", fontsize = "18")
title("Poisson, $\lambda = 10$")
plot(x,poisson.pmf(x,10),"--")
```

```python
x = linspace(0, 15, 16)
figure(figsize=(5, 3))
scatter(x,poisson.pmf(x,0.9))
ylabel("$f_X(x)$", fontsize = "18")
xlabel("$x$", fontsize = "18")
title("Poisson, $\lambda = 0.9$")
plot(x,poisson.pmf(x,0.9),"--")
```

####b. Normal

```python
x = linspace(0, 50, 200)
figure(figsize=(5, 3))
plot(x,norm.pdf(x,25,4))
ylabel("$f_X(x)$", fontsize = "18")
xlabel("$x$", fontsize = "18")
title("Normal, $\mu = 25$, $\sigma = 4$")
```

```python
x = linspace(-10, 10, 200)
figure(figsize=(5, 3))
plot(x,norm(0,1).pdf(x))
ylabel("$f_X(x)$", fontsize = "18")
xlabel("$x$", fontsize = "18")
title("Normal, $\mu = 0$, $\sigma = 1$")
```

#### c. Exponential

```python
x = linspace(0,9,100)
figure(figsize=(5, 3))
plot(x, expon.pdf(x, 0, 0.5))
ylabel("$f_X(x)$", fontsize = "18")
xlabel("$x$", fontsize = "18")
title("Exponential, $\lambda = 2$")
```

```python
x = linspace(0,10,100)
figure(figsize=(5, 3))
plot(x,expon.pdf(x,0,2))
ylabel("$f_X(x)$", fontsize = "18")
xlabel("$x$", fontsize = "18")
title("Exponential, $\lambda = 0.5$")
```
