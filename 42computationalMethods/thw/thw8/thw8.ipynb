```python
a = np.genfromtxt('quakes.csv', delimiter = ',')
```

```python
figure(figsize(20,10))
scatter(a[:,4],a[:,3], s = 2)
xlabel("Longitud", fontsize = 20)
ylabel("Latitud", fontsize = 20)
tick_params(labelsize = '15')
xlim(-180,180)
ylim(-80,80)
```

```python
h = hist(a[:,9], bins = 50, range=(1.999, 7), normed = True)
plt.title("Frecuencia relativa de la magnitud de los terremotos entre los anos 2000 y 2005", fontsize = 18)
plt.xlabel("magnitud", fontsize = 16)
plt.ylabel("frecuencia relativa", fontsize = 16)
tick_params(labelsize = '15')
```

```python
t = a[:,2]
t_ = zeros(len(a[:,2])-1)
for i in range(len(a[:,2])-1):
    t_[i] = t[i+1]-t[i]
```

```python
t_mean = mean(t_)
```

```python
figure(figsize(20,10))
plt.xlim(0, 10000)
plt.ylim(0, 0.0007)
xlabel("tiempo de espera /s", fontsize = 16)
ylabel("prob", fontsize = 16)
tick_params(labelsize = '15')
h = hist(t_, bins = 50, normed = True, range = (0,10000),label = "Datos Experimentales")
plot(h[1],(1./t_mean)*np.exp(-h[1]/t_mean),c = 'r', label = "Distribucion Exponencial", linewidth = 3)
legend(prop={'size':18})
```
