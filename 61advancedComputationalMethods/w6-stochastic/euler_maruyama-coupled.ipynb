dx_1 = x_2dt+alpha dW_1

dx_2 = -x_1dt+beta dW_2

alpha = beta = 0.1
x_1(0) = 1
x_2(0) = 0

```python
def step(t_old, x1_old, x2_old):
    
    t_new = t_old + dt
    x1_new = x1_old + x2_old*dt + alpha*np.sqrt(dt)*np.random.normal()
    x2_new = x2_old - x1_old*dt + beta*np.sqrt(dt)*np.random.normal()
    return t_new, x1_new, x2_new
```

```python
t_total = 10.0
alpha = 0.1
beta = 0.1
x1_0 = 1.0
x2_0 = 0.0
dt = 0.001
n_points = int(t_total/dt)
    
def steps():

    T = np.zeros(n_points)
    X1 = np.zeros(n_points)
    X2 = np.zeros(n_points)
    T[0] = 0
    X1[0] = x1_0
    X2[0] = x2_0
    
    for i in range(1,n_points):
    
        t_new, x1_new, x2_new = step(T[i-1],X1[i-1],X2[i-1])
        T[i] = t_new
        X1[i] = x1_new
        X2[i] = x2_new
    return T, X1, X2
```

```python
n_runs = 10

TT = np.zeros((n_points,n_runs))
XX1 = np.zeros((n_points,n_runs))
XX2 = np.zeros((n_points,n_runs))

for i in range(n_runs):
    T,X1,X2 = steps()
    TT[:,i] = T
    XX1[:,i] = X1
    XX2[:,i] = X2
```

```python
X1_ave = np.zeros(n_points)
X1_var = np.zeros(n_points)

X2_ave = np.zeros(n_points)
X2_var = np.zeros(n_points)

for i in range(n_points):
    X1_ave[i] = np.average(XX1[i,:])
    X1_var[i] = np.var(XX1[i,:])
    
    X2_ave[i] = np.average(XX2[i,:])
    X2_var[i] = np.var(XX2[i,:])
```
