##2. Parameters of the movements
The following lines are linux commands, which generates a list the 1000 files of the directory Brahe-3141-f/ and writes a file with them.
Then the C program hw4t.c is compiled and executed and it generates the list of parameters, saving it in the file parameters.dat, which has 5 columns labeled in the following order: $\theta$, $\phi$, $y_0$, $v_0$ and $g$.

```python
!ls Brahe-3141-f/ > names.dat
!cc hw4t.c -lm
!./a.out
```

Then we read the file parameters.dat with python and we plot $\theta$ vs. $g$. The following cell of code also sorts the array by $\theta$ value.

```python
parameters_ns = loadtxt('parameters.dat')

ind = lexsort((parameters_ns[:,4],parameters_ns[:,3],parameters_ns[:,2],parameters_ns[:,1],parameters_ns[:,0]))
parameters = parameters_ns[ind]

g = parameters[:,4]
theta = parameters[:,0]

figure(figsize(10,5))
xlabel(r'$\theta$', fontsize=20)
ylabel(r'$g(\theta)$', fontsize=20)
xlim([0,180])
scatter(theta,g, s=10)
```

##3. Find the dispersion $F$ of $g$
Next we generate and plot the function F representing the dispersion from the average And save it in f_data.dat, a file of two columns, the first is θ and the second F. The C scripts hw41.c and hw42.c will read that file.

```python
g_average = average(g)
F = 1.0 - g/g_average

savetxt('f_data.dat', column_stack((theta, F)), fmt='%f')

figure(figsize(10,5))
xlabel(r'$\theta$', fontsize=20)
ylabel(r'$F (\theta)$', fontsize=20)
xlim([0,180])
scatter(theta, F, s=10)
```

##4. Test the hyphoteses

###4.1 Hypothesis 1

Then we compile and run the C script hw41.c, which uses least squares to find the parameters $a_0$ and $a_1$ for the first hyphotesis

$$F = a_0\cos(2\theta)+a_1$$

```python
!cc hw41.c -lm
!./a.out
```

The next figure compares the measurements and the fit for this hypothesis.

```python
f1 = loadtxt('f1parameters.dat')
a0 = f1[0]
a1 = f1[1]
figure(figsize(10,5))
xlabel(r'$\theta$', fontsize=20)
ylabel(r'$F (\theta)$', fontsize=20)
xlim([0,180])
scatter(theta, F, s=10, label = 'Measurements')
plot(theta, a0*cos(2.0*theta*pi/180.0)+a1, label = 'Fit')
legend(fontsize=15)
```

###4.2 Hyphotesis 2

```python
f2 = loadtxt('f2parameters.dat')
b0 = f2[0]
b1 = f2[1]
b2 = f2[2]
figure(figsize(10,5))
xlabel(r'$\theta$', fontsize=20)
ylabel(r'$F(\theta)$', fontsize=20)
xlim([0,180])
scatter(theta, F, s=10, label = 'Measurements')
plot(theta, b0+b1*theta+b2*theta**2, label = 'Fit')
legend(fontsize=15)
```
